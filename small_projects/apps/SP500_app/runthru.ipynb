```python
import pandas as pd

def load_data():
    url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
    html = pd.read_html(url, header= 0)
    df = html[0]
    return df
```

```python
df = load_data()
df
```

```python
sector = df['GICS Sector']
len(sector)
```

```python
# aggregating the data 

sector = df.groupby('GICS Sector')
sector.first()
```

```python
sector.describe()
```

```python
sector.get_group('Health Care')
```

```python
import yfinance as yf 

data = yf.download(
    tickers = list(df.Symbol),
    period = '1y',
    interval= '1d',
    group_by= 'ticker',
    auto_adjust= True, 
    prepost= True, 
    threads = True,
    proxy = None
)
# wiki compatability issue, worked first time then failed second time with same code 
```
