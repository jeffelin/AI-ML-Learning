# train test split 

```python
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(df, test_size = 0.25)
```

```python
df_train.shape 
```

```python
# preparing our data for modeling 

X_train = df_train.drop(columns = ['target', 'target_name']).values 
y_train = df_train['target'].values 
```

```python
X_train, y_train
```

```python
# baseline model -> randomly guessing would yield an accuracy of 1/3 for equal datasets which is the worst

#simple manual model 

def single_feature_prediction(petal_length):
    if petal_length < 2.5: 
        return 0 
    elif petal_length < 4.8:
        return 1
    else: 
        return 2
```

```python
first_case = X_train[:,2 ]

manual_y_predictions = np.array([single_feature_prediction(val) for val in first_case])

manual_y_predictions == y_train #seeing true or false 

manual_model_accuracy = np.mean(manual_y_predictions == y_train)

print(f"Manual Model Accuracy: {manual_model_accuracy * 100:.2f}%")
```

```python
# modeling with logistic regression --> classification probabilities

from sklearn.linear_model import LogisticRegression

model = LogisticRegression() 


# never want to evaluate same data on training 
model.fit(X_train, y_train)
model.score(X_train, y_train)
```

```python
# validation test aka test set exc. cross validation 

Xt, Xv, yt, yv = train_test_split(X_train, y_train, test_size= 0.25)

y_pred = model.predict(Xv)
np.mean(y_pred == yv)
```

```python
model.score(Xv, yv) # short cut 
```

```python
# cross validation for each batch can be tested and iterated. averaging those testing cases is cv. 


from sklearn.model_selection import cross_val_score, cross_validate, cross_val_predict

model = LogisticRegression() 

accuracy = np.mean(cross_val_score(model, X_train, y_train, cv = 5, scoring = 'accuracy'))


accuracy
```

```python
# where are we misclassifying 

y_pred = cross_val_predict(model, X_train,  y_train, cv = 5) # trained on this data for the initial rows 
predicted_correctly_mask = y_pred == y_train
not_predicted_correctly = ~predicted_correctly_mask
```

```python
df_predictions = df_train.copy() 

df_predictions['correct_prediction'] = predicted_correctly_mask
df_predictions['prediction'] = y_pred 
df_predictions['prediction_label'] = df_predictions['prediction'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
df_predictions.head() 
```

```python
sns.scatterplot(x = 'petal length (cm)', y = 'petal width (cm)', hue = 'prediction_label', data=df_predictions)
```

```python
sns.scatterplot(x = 'petal length (cm)', y = 'petal width (cm)', hue = 'target_name', data=df_predictions) # seeing the comparison between predicted and true cases 
```

```python
def plot_incorrect_predictions(df_predictions, x_axis_feature, y_axis_feature_):
    fig, axs = plt.subplots(2,2, figsize = (10,10))
    axs = axs.flatten() 
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature_, hue ='prediction_label', data = df_predictions, ax = axs[0])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature_, hue ='target_name', data = df_predictions, ax = axs[1])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature_, hue ='correct_prediction', data = df_predictions, ax = axs[2])
    axs[3].set_visible

    plt.show()
```

```python
plot_incorrect_predictions(df_predictions, 'petal length (cm)', 'petal width (cm)')
```
